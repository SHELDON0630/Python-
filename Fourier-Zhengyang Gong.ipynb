### Introduction: This file mainly introduces the use of Fast Fourier Transform to discrete data points and use it the recognize the pitch of sample data. 
1.Import the basic functions

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

2. Define x and f(x) so that $f(x) = +1$ for $0 \leq x < 0.5$ and $-1$ for $0.5 \leq x < 1$ [1]

```python
x=np.linspace(0,1,1000)
f1=np.ones(500)       # define f1 an array of 500 1s. 
f2=np.linspace(-1,-1,500)  # define f2 as an array of 500 -1s. 
f=np.append(f1,f2)   # Add the two array together to create an array with 500 1 and 500 -1. 
```

3. Plot the graph of x and f(x) and the Fourier coefficient using FFT. 

```python
c_fft = np.fft.rfft(f) # transform the original data using Fast Fourier transform
                       # note the form of the call -
                       # we're using the numpy rfft routine in the numpy fft submodule
plt.figure()
plt.plot(x,f)
plt.title("Orignial data of x and f(x)")
plt.xlabel('x')
plt.ylabel('f(x)');
plt.figure()
plt.xlim(0,500)  #limit xaxis to 500 due to only 500 points
plt.plot(abs(c_fft))   # plot the absolute coefficients
plt.title("Fourier transform of square wave above ")
plt.xlabel('k')
plt.ylabel('coefficient $c_k$');
```

 4.  Define another square wave and plot the graph with its Fourier Transform coefficient with $0\leq x \leq 1000$, with $f(x) = +1$ for $0 \leq x \leq 10$[2]

```python
x1=np.linspace(0,1000,1000)
f1new=np.ones(10)
f2new=np.zeros(990)
fnew=np.append(f1new,f2new)
```

```python
c_fftnew = np.fft.rfft(fnew) # transform the original data using Fast Fourier transform
                       # note the form of the call -
                       # we're using the numpy rfft routine in the numpy fft submodule
plt.figure()
plt.plot(x1,fnew)
plt.title("Orignial data of x and f(x)")
plt.xlabel('x')
plt.ylabel('f(x)');
```

```python
plt.figure()
plt.xlim(0,500)#limit xaxis to 500 due to only 500 points
plt.plot((c_fftnew.real))   # use the real value 
plt.title("Fourier transform of square wave x and f(x) above ")
plt.xlabel('k')
plt.ylabel('coefficient $c_k$');
```

5. Import sample D and draw the grphs about sample D. Calculate the pitch of D using FFT. 

```python
sampleD = np.loadtxt("Sample_D.txt",float)
plt.figure()
plt.plot(sampleD[0:2000]) # Plot the first 2000 data points of D 
plt.title('First 2000 sample data D' )
plt.xlabel('sample points')
plt.ylabel('Frequency(Hz)')
```

```python
plt.figure()
plt.plot(sampleD[::500])
plt.title('Every 500 points sample data D' )
plt.xlabel('sample points')
plt.ylabel('Frequency(Hz)')
```

```python
c_fftD = np.fft.rfft(sampleD) # transform the original data using Fast Fourier Transform 
                       # note the form of the call -
                       # we're using the numpy rfft routine in the numpy fft submodule

plt.figure()
plt.plot(c_fftD[0:10000].real)   # plot the first 10000 real Fourier coefficients
plt.title("Fourier transform of 'sample D '")
plt.xlabel('k')
plt.ylabel('coefficient $c_k$');
```

```python
freD= 44.1*1000*590/len(sampleD)
print('The ferquency for sample D is', freD, 'Hz, with the fundamental k in the graph is',590,'which is C4' )
```

For Sample D, the Graph of the first 2000 points is a bit clearer than the others. The every 500th points graph shows a clear periodic characters. The fourier transfrom coefficient has relative less peak than the others. All the graphs are smooth.
