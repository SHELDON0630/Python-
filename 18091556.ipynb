# 1. Solving TISE for Infinite Square Well 
## 1.1 Define the Infinite Potential Square Well

```python
def V(x):
    '''
    This function is defined for creating an infinite potential square well with boundry (aneg,apos)
    Input is x: the x- position
    Output: The potential value of x-poisiton
    '''
    return 0.0
```

## 1.2 Define the Separate First-Order Equation for TISE 

$$ V_x = \begin{cases} 0 & \text{if } -a \leq x \leq +a, \\ 
\infty  & \text{if } |x| > a, \end{cases} [1]
$$

```python
def f(r,x,E): 
    '''
    This function is deifned as the two separate first order equations of TISE. 
    Inputs: r: with wavefunction value Ψ as r[0] and differentiaon value of the wavefunction ϕ as r[1]
    x: the x-poistion
    E: the energy value for the wavefunction 
    Output: the calculated value for fΨ and fϕ in an array 
    '''
    psi = r[0]  # The first element in r represent the wavefucntion value Ψ 
    phi = r[1]# second element of r represents the differntiation value of wavefunction ϕ 
    fpsi = phi   # ϕ is the derivative of Ψ with respect to x as difinition
    fphi = 2*m*(V(x)-E)*(psi)/((hbar)**2)  # the derivative of ϕ with respect to x is 2*m*Ψ*(V-E)/ħ^2
    return np.array([fpsi,fphi]) # convert fpsi and fphi to an array form
```

## 1.3 Define the Runge Kutta Fourth Order Equation for Solving f Function Above

```python
def RungeKutta2d(r,xpoints,E):
    '''Fourth-order Runge-Kutta rewritten for two simultaneous equations
    Inputs: r: 2-component array , first component is Ψ, second ϕ 
            xpoints: array of x-position values
            E: energy value of current wavefunction
    Outputs: [psipoints, phipoints]: solutions for Ψ(x) and ϕ(x), numpy arrays one longer than xpoints'''
    psipoints = [] # getting empty arrays ready for the vector variables
    phipoints = []
    for x in xpoints:
        psipoints.append(r[0])
        phipoints.append(r[1])
        k1 = h*f(r,x,E)
        k2 = h*f(r+0.5*k1, x+0.5*h,E)
        k3 = h*f(r+0.5*k2, x+0.5*h,E)
        k4 = h*f(r+k3, x+h,E)
        r = r + (k1 + 2*k2 + 2*k3 + k4)/6
    psipoints.append(r[0])  # Adding calculated elements to the arrays created 
    phipoints.append(r[1])
    return np.array([psipoints, phipoints])
```

## 1.4 Define the Width of Potential Well with xpoints Array and Set initial value for Ψ and ϕ

```python
#Define the initial value for r array 
psi_initial=0   # at thye boundry Ψ=0
phi_initial=5   # at the ϕ is setting randomly as 5
r=np.array([psi_initial,phi_initial])

# Define the width of the wall as (apos-aneg)= 10^(-10)m 
aneg = -5*10**(-11)         # The left boundry, in units of m 
apos = 5*10**(-11)          # The right boundry, in units of m 
N = 1000                    # number of points for Runge-Kutta
h = (apos - aneg)/N         # size of Runge-Kutta steps

# Define the x points array for Runge-Kutta from left boundry to right boundry with step of h 
xpoints = np.arange(aneg, apos, h)
```

## 1.5 Apply Secant Method to Find the Ground Energy Value for the Wavefunction

```python
# Define the initial guess for conditions
E1 = 0                      # guess 1 of initial E value for ground state, in units of J 
E2 = e                     # guess 2 of initial E value for ground state, in units of J

# Solve for the first guess
soln1 = RungeKutta2d(r,xpoints,E1) # the whole array of solutions for first energy guess 
psi1 = soln1[0,N]            # final element is at index N (counting from zero), which is the Ψ value at right boundry 

# solve for the second guess
psi2 = RungeKutta2d(r,xpoints,E2)[0,N]# final element for energy guess 2, which is the Ψ value at right boundry

#set the tolerance of secant method and run the loop to ensure Ψ=0 at right boundry 
tolerance =  e/1000                # set the tolerance for convergence
while abs(E2-E1) > tolerance: 
    E3 = E2 - psi2*(E2-E1)/(psi2-psi1)  
    # update guess of energy  for the next iteration
    E1 = E2 
    E2 = E3 
    # and Ψ is calculated with new guess
    psi1 = RungeKutta2d(r,xpoints,E1)[0,N]
    psi2 = RungeKutta2d(r,xpoints,E2)[0,N]
    
print("The ground state energy for this infinite square well numerically is ", E3, "J")

#Calcualte analytic ground state energy soluction:
En=(np.pi**2)*(hbar**2)/(2*m*((apos-aneg)**2))
print("The ground state energy for this infinite square well anlytically is ", En, "J")
```

## 1.6 Plot the Graph of Wavefunction for Infinite Square Well at Ground State

```python
psi_calculated = RungeKutta2d(r,xpoints,E3)[0]  # final psi calculated array with E3(fianl ground state energy)
# The array psi_calculated has one more elements than xpoints, a new x array has to be deifned to plot the graph
xplotpoints = np.arange(aneg, apos+h, h) # array with same properties as xpoints but includes end point

plt.figure()
plt.plot(xplotpoints,psi_calculated,label="Ψ(x) calculated ")
plt.legend(loc="best")
plt.axvline(x=aneg,c="k",ls="-", lw=2.5)# draw black dotted line at left boundry
plt.axvline(x=apos,c="k",ls="-", lw=2.5)
plt.xlabel("x-positison(m)")
plt.ylabel("Ψ ")
plt.title("Calculated wavefunction for infinite square well ");
```

## 1.7 Calculate and Plot the Normalised Wavefunction and Compare with Analytic Solutions

```python
# Calculate the integral of Ψ^2 using trapezoidal rule
Integral=sum((psi_calculated**2)*h) # f(a) and f(b) is 0 for Ψ at boundry and sum f(a+xh) is all the elements in psi array 

# Calculate the normalised wavefunction: Use the original function devided by square root of integral of Ψ^2
psi_normalised= psi_calculated/(Integral**0.5) 

#Caculate the analytic solution wavefunction at ground state 
Analytic=np.cos(np.pi*xplotpoints/(apos-aneg))/(apos**0.5)

plt.figure()
plt.plot(xplotpoints,psi_normalised,label="Ψ(x) normalised ")
plt.plot(xplotpoints,Analytic,'-.',label="Ψ(x) analytic solution ")
plt.legend(loc="best")
plt.axvline(x=aneg,c="k",ls="-", lw=2.5)# draw black dotted line at left boundry
plt.axvline(x=apos,c="k",ls="-", lw=2.5)
plt.xlabel("x-positison(m)")
plt.ylabel("Ψ ")
plt.title("Calculated wavefunction for infinite square well ");
```

## 1.8 Calculate and Plot the Normalised Wavefunction and Compare with Analytic Solutions for Different Energy State

```python
# define function Efinding of secant method same as above
def Efinding(E1,E2):
    '''
    This function is defined to use the secant method to find the energy in different state, 
    Input: E1:the first energy guess 
           E2:the second energy guess
    Output: The final energy value that satisfiy the boundry condition
    '''
    #solving for the first guess 
    psi1 = RungeKutta2d(r,xpoints,E1)[0,N] # final element is at index N (counting from zero), which is the Ψ value at right boundry 
    # solve for the second guess
    psi2 = RungeKutta2d(r,xpoints,E2)[0,N]# final element for energy guess 2, which is the Ψ value at right boundry
    #set the tolerance of secant method and run the loop to ensure Ψ=0 at right boundry 
    tolerance =  e/1000                # set the tolerance for convergence
    while abs(E2-E1) > tolerance: 
        E3 = E2 - psi2*(E2-E1)/(psi2-psi1)  
        # update guess of energy  for the next iteration
        E1 = E2 
        E2 = E3 
        # and Ψ is calcualted with new guess
        psi1 = RungeKutta2d(r,xpoints,E1)[0,N]
        psi2 = RungeKutta2d(r,xpoints,E2)[0,N]
    return E3
```

$$
where $V_0=1000eV$,$a=5\times10^{-11}m$ and outside the well $V=0$ for $|x| > a$

```python
def V_finite(x):
    '''
    This function is defined for creating an finite potential square well with boundry (aneg,apos)
    Input is x: the x- position
    Output: The potential value of x-poisiton
    '''
    if abs(x)<apos/2:  # when x is outside the well, the potential is 600e
        V=0
    elif abs(x)>apos/2: # if x position is within the well, potential is zero
        V=1000*e
    return V
```

## 2.2 Define the Separate First-Order Equation for TISE 

```python
# This is the same function as f(r,x,E) above , but use V_fintie(x) instead of V(x) instead
def f_finite(r,x,E): 
    '''
    This function is deifned as the two separate first order equations of TISE. 
    Inputs: r: with wavefunction value Ψ as r[0] and differentiaon value of the wavefunction ϕ as r[1]
    x: the x-poistion
    E: the energy value for the wavefunction 
    Output: the calculated value for fΨ and fϕ in an array 
    '''
    psi = r[0]  # The first element in r represent the wavefucntion value Ψ 
    phi = r[1]# second element of r represents the differntiation value of wavefunction ϕ 
    fpsi = phi   # ϕ is the derivative of Ψ with respect to x as difinition
    fphi = 2*m*(V_finite(x)-E)*(psi)/((hbar)**2)  # the derivative of ϕ with respect to x is 2*m*Ψ*(V-E)/ħ^2
    return np.array([fpsi,fphi]) # convert fpsi and fphi to an array form
```

## 2.3 Define the Runge Kutta Fourth Order Equation for Solving f_finite Function Above

```python
def RungeKutta2d_finite(r,xpoints,E):
    '''Fourth-order Runge-Kutta rewritten for two simultaneous equations
    Inputs: r: 2-component array , first component is Ψ, second ϕ 
            xpoints: array of x-position values
            E: energy value of current wavefunction
    Outputs: [psipoints, phipoints]: solutions for Ψ(x) and ϕ(x), numpy arrays one longer than xpoints'''
    psipoints = [] # getting empty arrays ready for the vector variables
    phipoints = []
    for x in xpoints:
        psipoints.append(r[0])
        phipoints.append(r[1])
        k1 = h*f_finite(r,x,E)
        k2 = h*f_finite(r+0.5*k1, x+0.5*h,E)
        k3 = h*f_finite(r+0.5*k2, x+0.5*h,E)
        k4 = h*f_finite(r+k3, x+h,E)
        r = r + (k1 + 2*k2 + 2*k3 + k4)/6
    psipoints.append(r[0])  # Adding calculated elements to the arrays created 
    phipoints.append(r[1])
    return np.array([psipoints, phipoints])
```

## 2.4 Define a Function to Find the Energy Eigenvalues for Different n

```python
# define function Efinding_finte the same seacnt method as above. But now Ψ outside the well is needed. 
def Efinding_finite(E1,E2):
    '''
    This function is defined to use the secant method to find the energy in different state, for larger range than the well itself 
    Input: E1:the first energy guess 
           E2:the second energy guess
    Output: The final energy value that satisfiy the boundry condition
    '''
    #solving for the first guess 
    psi1 = RungeKutta2d_finite(r,xpoints,E1)[0,N]
    # N is the element element is Ψ at right boundry which is x=apos for E1
    # solve for the second guess
    psi2 = RungeKutta2d_finite(r,xpoints,E2)[0,N]# Nis the element element is Ψ at right boundry which is x=apos
    #set the tolerance of secant method and run the loop to ensure Ψ=0 at right boundry for E2
    tolerance =  e/1000                # set the tolerance for convergence
    while abs(E2-E1) > tolerance: 
        E3 = E2 - (psi2)*(E2-E1)/(psi2-psi1)   
        # update guess of energy  for the next iteration
        E1 = E2 
        E2 = E3 
        # and Ψ is calcualted with new guess
        psi1 = RungeKutta2d_finite(r,xpoints,E1)[0,N]
        psi2 = RungeKutta2d_finite(r,xpoints,E2)[0,N]
        
    return E3
```

# 4. Construct An Unsquare Infinite Well and Plot Its Wavefunctions and Energies for Different n
$$
V_x = \begin{cases} V_0 cos(\frac{x}{a}) & \text{if }  0<x<a  , \\ 
    -V_0ln\frac{|x|}{a}  & \text{if } -a/2<x<0,\\
    V_0 sin(\frac{|x|}{a})  & \text{if } -a<x<-a/2,\\
    0 & \text{if } |x| > a/2
    \end{cases} 
$$

# 5. Solving TISE for Another Potential
## 5.1 Define the Potential 

$$
V_x = \begin{cases} V_0 \frac{a-x}{a} & \text{if } 0\leq x  , \\ 
    V_0 \frac{|x|}{a}  & \text{if } x<0 \end{cases} 
$$
where $V_0=1000eV$,$a=5\times10^{-11}m$

```python
def V_my(x):
    '''
    This function is defined for creating a linear potential at either side of y-axis
    Input is x: the x- position
    Output: The potential value of x-poisiton
    '''
    if x>0:
        V=abs(1000*e*(apos-x)/apos)
    else:
        V=1000*e*abs(x)/apos
    return V
```

## 5.2 Define the Separate First-Order Equation for TISE 

```python
# This is the same function as f(r,x,E) above , but use V_my(x) instead of V(x) instead
def f_my(r,x,E): 
    '''
    This function is deifned as the two separate first order equations of TISE. 
    Inputs: r: with wavefunction value Ψ as r[0] and differentiaon value of the wavefunction ϕ as r[1]
    x: the x-poistion
    E: the energy value for the wavefunction 
    Output: the calculated value for fΨ and fϕ in an array 
    '''
    psi = r[0]  # The first element in r represent the wavefucntion value Ψ 
    phi = r[1]# second element of r represents the differntiation value of wavefunction ϕ 
    fpsi = phi   # ϕ is the derivative of Ψ with respect to x as difinition
    fphi = 2*m*(V_my(x)-E)*(psi)/((hbar)**2)  # the derivative of ϕ with respect to x is 2*m*Ψ*(V-E)/ħ^2
    return np.array([fpsi,fphi]) # convert fpsi and fphi to an array form
```

## 5.3 Define the Runge Kutta Fourth Order Equation for Solving f_my Function Above

```python
def RungeKutta2d_my(r,xpoints,E):
    '''Fourth-order Runge-Kutta rewritten for two simultaneous equations
    Inputs: r: 2-component array , first component is Ψ, second ϕ 
            xpoints: array of x-position values
            E: energy value of current wavefunction
    Outputs: [psipoints, phipoints]: solutions for Ψ(x) and ϕ(x), numpy arrays one longer than xpoints'''
    psipoints = [] # getting empty arrays ready for the vector variables
    phipoints = []
    for x in xpoints:
        psipoints.append(r[0])
        phipoints.append(r[1])
        k1 = h*f_my(r,x,E)
        k2 = h*f_my(r+0.5*k1, x+0.5*h,E)
        k3 = h*f_my(r+0.5*k2, x+0.5*h,E)
        k4 = h*f_my(r+k3, x+h,E)
        r = r + (k1 + 2*k2 + 2*k3 + k4)/6
    psipoints.append(r[0])  # Adding calculated elements to the arrays created 
    phipoints.append(r[1])
    return np.array([psipoints, phipoints])
```

## 5.4 Define a Function to Find the Energy Eigenvalues for Different n

```python
# define function Efinding_my the same seacnt method as above. But there is no well of Ψ.  
#Instead of using the first and last element of Ψ value, we choose 0.5*N and 1.5*N(Ψ value at aneg and apos) to calculate. 
def Efinding_my(E1,E2):
    '''
    This function is defined to use the secant method to find the energy in different state, for larger range than the well itself 
    Input: E1:the first energy guess 
           E2:the second energy guess
    Output: The final energy value that satisfiy the boundry condition
    '''
    #solving for the first guess 
    psi1 = RungeKutta2d_my(r,xpoints,E1)[0,int(N)]
    # N is the element element is Ψ at right boundry which is x=apos for E1
    # solve for the second guess
    psi2 = RungeKutta2d_my(r,xpoints,E2)[0,int(N)]# N is the element element is Ψ at right boundry which is x=apos
    #set the tolerance of secant method and run the loop to ensure Ψ=0 at right boundry for E2
    tolerance =  e/1000                # set the tolerance for convergence
    while abs(E2-E1) > tolerance: 
        E3 = E2 - (psi2)*(E2-E1)/(psi2-psi1)   
        # instead to find Ψ=0 at right boundry we use Ψ(x=apos)-Ψ(x=aneg)=0 to satisfy the boundry condtion
        # update guess of energy  for the next iteration
        E1 = E2 
        E2 = E3 
        # and Ψ is calcualted with new guess
        psi1 = RungeKutta2d_my(r,xpoints,E1)[0,int(N)]
        psi2 = RungeKutta2d_my(r,xpoints,E2)[0,int(N)]
    return E3
```
